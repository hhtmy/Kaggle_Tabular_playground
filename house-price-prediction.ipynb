```python
df=pd.concat((train_df.iloc[:,1:-1],test_df.iloc[:,1:]))
```

```python
num_df=df.dtypes[df.dtypes!='object'].index
```

```python
from scipy.stats import skew
cols=df[num_df].apply(lambda x:skew(x))
cols=cols[cols>0.5].index
df[cols]=np.log1p(df[cols])
df[num_df].skew()
```

```python
df[num_df]=df[num_df].fillna(0)
```

```python
df=pd.get_dummies(df,dummy_na=True)
```

```python
train=train_df.shape[0]
train_set=np.array(df[:train].values,dtype=np.float32)
test_set=np.array(df[train:].values,dtype=np.float32)
```

```python
y=np.array(train_df.SalePrice.values.reshape(-1,1),dtype=np.float32)
y
```

```python
from sklearn.ensemble import GradientBoostingRegressor
reg=GradientBoostingRegressor(max_depth=2,n_estimators=120)
reg.fit(train_set,y)
y_pred=reg.predict(test_set)
```

```python
test_df['SalePrice']=pd.Series(y_pred.reshape(1,-1)[0])
submission=pd.concat([test_df['Id'],test_df['SalePrice']],axis=1)
submission.to_csv('submission.csv',index=False)
```
